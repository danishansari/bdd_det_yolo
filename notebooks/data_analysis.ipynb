## Detection (Yolov5) on BDD-100k-Dataset

BDD-100k-dataset on various task of detection, segmentation Pose estimation, lane marking etc. 
In this experiment we're only interested in the detection task with following classes.
  * 0: person
  * 1: rider
  * 2: car
  * 3: bus
  * 4: truck
  * 5: bike
  * 6: motor
  * 7: train
  * 8: traffic sign
  * 9: traffic light

## Datset exploration and analysis

### 1. Data distibution - train/eval sample split

```python
from PIL import Image
from matplotlib import pyplot as plt

im = Image.open("../plots/train_val_dist.png")
plt.axis("off")
plt.imshow(im)
```

From above graph we can observe that train-val ratio is 1/7, about 14%. Which is ok in terms of train-val split in most experiments, having a 20% split is considered ideal.

### 2. Classwise Data distribution

```python
from PIL import Image
from matplotlib import pyplot as plt

im1 = Image.open("../plots/training_class_dist.png")
im2 = Image.open("../plots/eval_class_dist.png")

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 8))
axes[0].set_axis_off()
axes[0].imshow(im1)
axes[1].set_axis_off()
axes[1].imshow(im2)
```

From above plot its evident that there is huge imbalance in the dataset, `car` being the majority and objects like `rider`, `bus`, `motor`, `bike`, `truck` are very less with `train` being the least of all samples.

### 3. Objects width and height distribution

```python
from PIL import Image
from matplotlib import pyplot as plt

im = Image.open("../plots/training_size_dist.png")
plt.axis("off")
plt.imshow(im)
```

From above plot we can see most of the object sizes are in the middle of the plor around `700x300` dimension.

### 4. Area Vs no of smaples in the dataset
